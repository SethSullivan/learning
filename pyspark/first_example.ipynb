```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("FirstExample").getOrCreate()

df = spark.read.csv("./data/GiveMeSomeCredit/cs-training.csv", header=True, inferSchema=True)
# Replace na with means 
df = df.na.fill(df.agg({"MonthlyIncome": "mean"}).first().asDict())
display(df.show(5))
```

```python
print(df.schema)
```

```python
df.groupBy("age").count().orderBy("age", ascending=False)
```
